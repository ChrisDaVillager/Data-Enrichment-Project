```python
# imports 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os, json
```

```python
df1 = pd.read_csv('Data/final_tmdb_data_2000.csv.gz')
df1.head()
```

```python
df1.duplicated().sum().sum()
```

```python
df2 = pd.read_csv('Data/final_tmdb_data_2001.csv.gz')
df2.head()
```

```python
df2.info()
```

```python
df2.duplicated().sum().sum()
```

```python
# concatentate dataframes
merge = [df1,df2]
merged_df = pd.concat(merge)
merged_df.head()
```

```python
merged_df.info()
```

```python
merged_eda = merged_df.copy()
merged_eda.dropna(inplace=True)
```

```python
merged_eda.isna().sum().sum()
```

```python
# checking budget values
merged_eda['budget'].value_counts()
```

```python
# checking revenue values
merged_eda['revenue'].value_counts()
```

**1. How many movies had at least some valid financial information (values > 0 for budget OR revenue)?**

```python
merged_eda['budget'].value_counts().plot(kind='bar');
```

```python
merged_eda['revenue'].value_counts().plot(kind='bar');
```

**2. How many movies are there in each of the certification categories (G/PG/PG-13/R)?**

```python
merged_eda['certification'].value_counts().plot(kind='bar');
```

**3. What is the average revenue per certification category?**

```python
sns.barplot(data=merged_eda, x='certification',y='revenue',ci=None);
```

**4. What is the average budget per certification category?**

```python
sns.barplot(data=merged_eda, x='certification',y='budget',ci=None);
```
